```python
data = pd.read_csv('../../data/titanic.csv', index_col='PassengerId')
data['Sex'] = data['Sex'].map({'female': 0, 'male': 1})
```

```python
data.isna().sum().sum()
```

```python
data.dropna(subset=['Pclass', 'Fare', 'Age', 'Sex'], axis=0, inplace=True)
data.isna().sum().sum()
```

```python
X = data[['Pclass', 'Fare', 'Age', 'Sex']]
y = data['Survived']
X.isna().sum().sum()
```

```python
clf = DecisionTreeClassifier(random_state=241)
clf = clf.fit(X, y)
```

```python
print(X.columns)
print(clf.feature_importances_)
```
