```python
import re
```

```python
women_names = data[data.Sex == 'female'].Name
```

```python
def get_best_women_names(names):
    str_names = ' '.join(names)
    str_names = re.findall(r"[\w']+", str_names)
    
    names_d = {}
    for name in str_names:
        if name in names_d.keys():
            names_d[name] += 1
        else:
            names_d[name] = 1
            
    # sort them
    names_d = dict(sorted(names_d.items(), key=lambda item: item[1], reverse=True))
    names_d = list(names_d.items())[:10]
    return names_d
```

```python
names_d = get_best_women_names(women_names)
names_d[2:]
```
